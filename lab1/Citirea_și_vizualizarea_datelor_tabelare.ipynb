# Heatmaps

Putem folosi hărți de căldură pentru a vizualiza corelația dintre variabile.

Acestea pot dezvălui clustere sau corelații.


```python
plt.figure(figsize= (12,12))
sns.heatmap(df.corr(),cmap='coolwarm', annot = True)
plt.show()
```

## Coeficienții de corelație:

- **1**: pe măsură ce o variabilă crește, cealaltă variabilă crește proporțional.
- **-1**: pe măsură ce o variabilă crește, cealaltă scade proporțional.
- **0**: nu există o relație liniară între variabile.

Putem observa că cea mai mare corelație pozitivă (0.54) este între vârstă și sarcini, sugerând că, pe măsură ce vârsta crește, numărul de sarcini tinde să crească și el.

Glucoza și Outcome arată o corelație pozitivă de 0.47, indicând că nivelurile mai mari de glucoză sunt asociate cu o probabilitate mai mare de diabet.

De asemenea, putem observa o corelație între insulină și grosimea pielii, implicând că nivelurile mai mari de insulină tind să fie asociate cu o piele mai groasă, ceea ce este menționat și în literatura de specialitate: [PubMed](https://pubmed.ncbi.nlm.nih.gov/2721339/).

# Valori lipsa

În seturile de date, aceste valori lipsă pot apărea sub diferite forme: „0”, „NA”, „NaN”, „NULL”, „Not Applicable” sau „None”. În acest caz, observăm că datele lipsă sunt marcate cu 0. Nimeni nu poate avea un BMI egal cu 0.


```python
missing_values = (df[['SkinThickness', 'BloodPressure', 'BMI']] == 0).sum()
print("Number of missing values:")
print(missing_values)
```

```python
mean_skin_thickness = df['SkinThickness'][df['SkinThickness'] != 0].mean()
df['SkinThickness'] = df['SkinThickness'].replace(0, mean_skin_thickness)
missing_values = (df[['SkinThickness', 'BloodPressure', 'BMI']] == 0).sum()
print("Number of missing values:")
print(missing_values)
```

Mean Imputation: este utilă atunci când datele urmează o distribuție normală; media reprezintă bine tendința centrală.

Median Imputation: este mai puțin sensibilă la valorile extreme (outliers).

Mode Imputation: este utilă pentru date categorice sau pentru date numerice discrete. Considerăm valoarea care apare cel mai frecvent într-un set de date. Această metodă este folosită pentru a înlocui valorile lipsă cu cea mai comună valoare dintr-o coloană dată.

Forward Fill: înlocuiește valorile lipsă cu ultima valoare observată din coloană.

Backward Fill: înlocuiește valorile lipsă cu următoarea valoare observată din coloană.

În setul de date anterior aveam doar date numerice, însă în acest set de date avem și date categorice. Deoarece algoritmii de învățare automată pot procesa doar valori numerice, este necesar să convertim aceste feature-uri categorice în reprezentări numerice corespunzătoare.

Pentru a face acest lucru, putem folosi următoarele metode:


Label Encoder: se atribuie fiecărei etichete o valoare numerică între 0 și (n_classes - 1).
 


One-Hot Encoder: fiecare feature este reprezentat printr-un vector binar în care o singură poziție este 1 (corespunzătoare categoriei respective), iar restul sunt 0.


```python
stroke_df.head()
```

```python
stroke_one_hot_df = pd.get_dummies(stroke_df)
stroke_one_hot_df.head()
```

```python
categorical_columns = stroke_df.select_dtypes(include=['object', 'category']).columns.tolist()
categorical_columns
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
label_mappings = {}

for column in categorical_columns:
    stroke_df[column] = label_encoder.fit_transform(stroke_df[column])
    label_mappings[column] = dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))

print(label_mappings)
stroke_df.head()
```

3. Numărați și afișați câte intrări lipsesc pentru caracteristica BMI și completați valorile lipsă. Justificați metoda aleasă.

```python
#TO DO
col_bmi = 'bmi'
assert col_bmi in stroke_df.columns, f"Nu găsesc coloana {col_bmi}!"

n_missing = stroke_df[col_bmi].isna().sum()
print(f"Valori lipsă în {col_bmi}: {n_missing}")

age_bins = [0, 18, 30, 45, 60, 75, 120]
age_labels = ['0-17','18-29','30-44','45-59','60-74','75+']
stroke_df['_age_bin'] = pd.cut(stroke_df['age'], bins=age_bins, labels=age_labels, right=False)

median_by_bin = stroke_df.groupby('_age_bin')[col_bmi].median()
print("Mediana BMI pe grupe de vârstă:\n", median_by_bin)

def fill_bmi(row):
    if pd.isna(row[col_bmi]):
        return median_by_bin.loc[row['_age_bin']]
    return row[col_bmi]

stroke_df[col_bmi] = stroke_df.apply(fill_bmi, axis=1)
stroke_df.drop(columns=['_age_bin'], inplace=True)

print(f"Valori lipsă rămase în {col_bmi}: {stroke_df[col_bmi].isna().sum()}")
```
